```python
from sympy import * 
import math
###
def str_to_fun(str):
    n=len(str)
    lst=[]
    for i in range(0,n,2):
        lst.append(str[i:i+2])  #
    len_lst=len(lst)   #a
    len_bit=int(math.log(len_lst,2))  #S  
    truth_table_input=[]      #S
    truth_table_output=[]
    for i in range(len_lst):
        truth_table_input.append(bin(i)[2:].zfill(len_bit))  #ilen_bit
        truth_table_output.append(bin(int(lst[i],16))[2:].zfill(len_bit))   #sbox_arrlen_bit
    fun_expr=[]     #
    x= IndexedBase('x')  #x
    for i in range(len_bit):
        f=0
        for j in range(len_lst):
            item=1    #2**len_bit-1item 1
            if truth_table_output[j][i]=='1':
                for k in range(len_bit):
                    if truth_table_input[j][k]=='0':
                        item*=(1+x[k])  #
                    else:
                        item*=x[k]
            f+=item   #
        f=Poly(f, domain='GF(2)').args[0]   #2  
        fun_expr.append(f)
    return fun_expr





from sympy import * 
import math
import random
x= IndexedBase('x')
len_bit=32
lst_sub=[]
for i in range(len_bit):       #
    lst_sub.append((x[i]**2,x[i]))      #
  

###Toffoli
def toffoli_f(fun1,fun2,fun3):
    g=expand(fun3+fun1*fun2)  #
    g=g.subs(lst_sub)      #
    g=Poly(g, domain='GF(2)')
    return g.args[0]

###CNOT
def cnot_f(fun1,fun2):
    g=expand(fun1+fun2)
    g=g.subs(lst_sub)
    g=Poly(g, domain='GF(2)')
    return g.args[0]

###(S)
def degree_rev(f):
    len_f=len(f)      #S    
    degree=Poly(f[0]).total_degree()   #  
    for i in range(1,len_f):                
        if Poly(f[i]).total_degree() > degree:
            degree=Poly(f[i]).total_degree()
    return degree

###
def high_degree_item(fun):
    degree=Poly(fun).total_degree()
    high_degree_item_set=set()
    for i in fun.args:
        if len(i.args)==degree:
            high_degree_item_set.add(i)
    return high_degree_item_set

###
def high_item_rev(f):
    len_f=len(f)
    degree=degree_rev(f)
    high_item_set=set()    #
    for i in range(len_f):
        for j in Poly(f[i]).args[0].args:
            if len(j.args)==degree:
                high_item_set.add(j)
    return high_item_set

###,count_lastcount_now
def gate_restore_lst(gate_lst,count_last,count_now):
    if count_last==-1:
        for i in gate_lst[count_now:0:-1]:
            gate_lst.append(i)
        gate_lst.append(gate_lst[0])
    else:
        for i in gate_lst[count_now:count_last:-1]:
            gate_lst.append(i)
    return gate_lst

###
def choice_lst_k(len_bit,k):
    lst=[]
    for i in range(len_bit):
        for j in range(len_bit):
            if j!=i and j!=k and i!=k:
                lst.append([k,i,j])
    return lst



###
def inverse(A):
    A_inverse = np.linalg.inv(A)   #A
    A_inverse = np.abs(A_inverse) %2   #2
    A_inverse = A_inverse.astype(int)   
    return A_inverse

###
def search_item_lst(fun,high_item_lst):
    lst=[]
    for i in range(len(high_item_lst)):
        lst.append(0)
    for i in fun.args:
        if i in high_item_lst:
            lst[high_item_lst.index(i)]=1
    return lst

###
def multi_bA(b,A):    
    lst=[]
    for i in range(len(b)):
        sum=0
        for j in range(len(b)):
            sum=sum+b[j]*A[j][i]
        lst.append(sum % 2)
    return lst

###cnotf
def add_cnot(high_mat,high_index,f,gate_lst,high_item_lst,target_f,num_qubit):
    len_f=len(f)
    for i in range(len_f):
        fun_lst=search_item_lst(f[i],high_item_lst)
        solve_lst=multi_bA(fun_lst,inverse(high_mat))   
        for j in range(len(solve_lst)):
            if solve_lst[j]==1:
                gate_lst.insert(high_index[j]+1,[gate_lst[high_index[j]][2],num_qubit+i])                                   
                f[i]=cnot_f(target_f[high_index[j]],f[i])
                target_f.insert(high_index[j]+1,0) ###0
                for k in range(j+1,len(high_index)):
                    high_index[k]=high_index[k]+1
    return gate_lst,f

###Toffoli
def gate_lst_opt(gate_lst):
    gate_len=len(gate_lst)
    i=0
    while i<gate_len-1:
        if gate_lst[i]==gate_lst[i+1]:
            gate_lst.pop(i)   #
            gate_lst.pop(i)  #
            gate_len=gate_len-2
            if (i-1)!=-1:
                i=i-1
        else:
            i=i+1   #
    return gate_lst

###
def gate_lst_res(gate_lst):
    num_t,num_c,num_n=0,0,0
    for i in gate_lst:
        if len(i)==3:
            num_t=num_t+1
        elif len(i)==2:
            num_c=num_c+1
        else:
            num_n=num_n+1
    dict={'ccx':num_t,'cx':num_c,'x':num_n}
    return dict




from sympy import *  
import math
import random
x= IndexedBase('x')

### ,
def syn_mon(mon,num_qubit):
    gate_lst=[]    #
    lst_cmp=[]     #
    lst_mon=list(mon.args)  #   
    current_f=[]   #qubit
    target_f=[]    #
    for i in range(num_qubit):
        current_f.append(x[i])  # 
    lst_x=[]                    # 
    for i in range(num_qubit):
        lst_x.append(x[i])
    for i in lst_x:
        if i not in lst_mon:
            lst_cmp.append(i)   #
    i,j=lst_x.index(lst_mon[0]),lst_x.index(lst_mon[1])
    k=lst_x.index(lst_cmp[0])
    lst_cmp.pop(0)   #
    gate_lst.append([i,j,k])
    current_f[k]=toffoli_f(current_f[i],current_f[j],current_f[k])
    target_f.append(current_f[k])
    lst_cmp.append(lst_mon[0])
    lst_cmp.append(lst_mon[1])
    lst_mon.pop(0) #0
    lst_mon.pop(0) #1   
    while lst_mon !=[]:   #lst_mon
        i,j,k=k,lst_x.index(lst_mon[0]),lst_x.index(lst_cmp[0])
        gate_lst.append([i,j,k])
        current_f[k]=toffoli_f(current_f[i],current_f[j],current_f[k])
        target_f.append(current_f[k])
        lst_cmp.append(lst_mon[0])
        lst_cmp.pop(0)
        lst_mon.pop(0)            
    return gate_lst,current_f,target_f





###2
def syn_low(f,num_qubit):
    len_f=len(f)   
    x= IndexedBase('x')
    lst_x=[]
    for i in range(num_qubit):
        lst_x.append(x[i])
    lst=[]
    K=10000
    lst_K=[]
    for i in range(1,len_f):
        for j in range(i):
            for k in range(len_f):
                for l in f[k].args:
                    if l==x[i]*x[j]:
                        lst.append([j,i,k+num_qubit])
                        K=k
                        break
                if K!=10000:
                    break
            for k in range(K+1,len_f):
                for l in f[k].args:
                    if l==x[i]*x[j]:
                        lst_K.append(k)
            reversed_lst_K = lst_K[::-1]
            for i in reversed_lst_K:
                lst.insert(-1,[K+num_qubit,i+num_qubit])
            for i in lst_K:
                lst.append([K+num_qubit,i+num_qubit])
            lst_K.clear()
            K=10000
    for i in range(len_f):
        for j in range(len_f):
            for k in f[j].args:
                if k==x[i]:
                    lst.append([i,j+num_qubit])
    for i in range(len_f):
        for j in f[i].args:
            if j==1:
                lst.append([i+num_qubit])
    return lst





from sympy import * 
import math
import random
import copy
x= IndexedBase('x')

###
def multi_step_syn(f,num_qubit):
    len_f=len(f)      #S    
    degree=degree_rev(f)   #
    gate_lst=[]          #
    while degree>2:
        high_item_set=high_item_rev(f)  #f
        high_item_lst=list(high_item_set)  #f
        for x in high_item_lst:
            lst,current_f,target_f=syn_mon(x,num_qubit)
            k=lst[-1][2]   #
            gate_lst.extend(lst)
            for i in range(len_f):
                if x in f[i].args:   #Xf[i]
                    gate_lst.append([k,i+num_qubit])
                    f[i]=cnot_f(current_f[k],f[i])    #syn_monqubit,
            gate_lst.extend(reversed(lst))   #1st
        degree=degree_rev(f)                    
    lst2=syn_low(f,num_qubit)
    gate_lst.extend(lst2)
    gate_lst=gate_lst_opt(gate_lst)
    return gate_lst




import numpy as np
from qiskit import *
from qiskit_aer import AerSimulator
num_qubit=5
len_f=5
q1=QuantumRegister(num_qubit,'a')
q2=QuantumRegister(len_f,'b')
c1 = ClassicalRegister(len_f)

qc=QuantumCircuit(q1,q2,c1)

def lst_to_qc(lst,num_qubit):
    for lst1 in lst:
        if len(lst1)==3:
            if lst1[2]<num_qubit:
                qc.ccx(q1[lst1[0]],q1[lst1[1]],q1[lst1[2]])
            else:
                qc.ccx(q1[lst1[0]],q1[lst1[1]],q2[lst1[2]-num_qubit])
        elif len(lst1)==2:
            if lst1[1]<num_qubit:
                qc.cx(q1[lst1[0]],q1[lst1[1]])
            elif lst1[1]>=num_qubit and lst1[0]<num_qubit:
                qc.cx(q1[lst1[0]],q2[lst1[1]-num_qubit])
            else:
                qc.cx(q2[lst1[0]-num_qubit],q2[lst1[1]-num_qubit])
        else:
            qc.x(q2[lst1[0]-num_qubit])

##
# qc.x(q1[0])
# qc.x(q1[1])
# qc.x(q1[2])
# qc.x(q1[3])
# qc.x(q1[4])

##
s='040B1F141A1509021B0508121D03061C1E13070E000D1118100C0119160A0F17'
f=str_to_fun(s)
print(f)

lst=multi_step_syn(f,num_qubit)
lst_to_qc(lst,num_qubit)
for i in range(len_f):
    qc.measure(q2[i],c1[i])

qc.draw('mpl')
```

```python
###i,j,k
def eval_fun(i,j,k,current_f,current_item_set,high_item_set,num_qubit,degree):
    len_f=len(current_f)
    m=math.factorial(num_qubit)          #i,j,k
    fun=toffoli_f(current_f[i],current_f[j],current_f[k])
    degree_fun=Poly(fun).total_degree()   #fun
    
    if degree_fun>degree:
        return 0      #
    elif degree_fun==degree:
        s=high_degree_item(fun)    #
        if degree_fun==degree and (len(s.difference(current_item_set))==1) and (s.issubset(high_item_set)):   
            return m    # s current_item_set  1 s high_item_set 
        else:
            return 0       #
    else:
        s=high_degree_item(fun) 
        sum=0
        for i in s:
            for j in high_item_set:
                if set(i.args).issubset(set(j.args)):   #i,j
                    sum=sum+len(i.args)**2
        return sum




from sympy import * 
import math
import random
import copy
x= IndexedBase('x')

###3
def elim_onedegree(f,num_qubit):
    len_f=len(f)      #    
    m=math.factorial(num_qubit)      #
    degree=degree_rev(f)   #
    gate_lst=[]          #
    high_index=[]      #    
    high_mat=[]        #
    target_f=[]        #
    current_item_set=set()   #
    high_item_set=high_item_rev(f)  #f   
    high_item_lst=list(high_item_set)  #f   
    len_high=len(high_item_set)                    #
    
    x=high_item_lst[0]
    lst,current_f,target_f=syn_mon(x,num_qubit)    
    current_item_set.add(x)
    gate_lst.extend(lst)    
    high_mat.append(search_item_lst(x+1,high_item_lst))  #search_item_lstx+1   
    count_last=-1    #   
    count_now=degree-2   #
    high_index.append(count_now)
    k=gate_lst[-1][2]
            
    while True:   
        choice_lst=choice_lst_k(num_qubit,k)  #choice_lst
        max_val=0   #0
        t=len(gate_lst)   
        if t<num_qubit:
            win_lst=gate_lst
        else:
            win_lst=gate_lst[t-num_qubit:t]  #    
        for lst in choice_lst:          #i,j,k
            if lst in win_lst:
                val=0
            else:   #
                val=eval_fun(lst[0],lst[1],lst[2],current_f,current_item_set,high_item_set,num_qubit,degree)
            if val>max_val:
                max_val=val   #
                max_ijk=lst   #
            if max_val==m:
                break
        if max_val==m:   
            gate_lst.append(max_ijk)
            count_now=count_now+1     #+1
            i,j,k=max_ijk[0],max_ijk[1],max_ijk[2]
            current_f[k]=toffoli_f(current_f[i],current_f[j],current_f[k])   #Toffoli
            target_f.append(current_f[k])
            high_mat.append(search_item_lst(current_f[k],high_item_lst))
            high_index.append(count_now)     #
            current_item_set=current_item_set.union(high_degree_item(current_f[k]))   #
            if len(current_item_set)==len_high:    #
                gate_lst=gate_restore_lst(gate_lst,count_last,count_now)
                gate_lst,f=add_cnot(high_mat,high_index,f,gate_lst,high_item_lst,target_f,num_qubit)
                gate_lst=gate_lst_opt(gate_lst)      #
                return gate_lst,f             #
        elif max_val==0:     #
            gate_lst=gate_restore_lst(gate_lst,count_last,count_now)       #    
            for i in range(count_now-count_last):
                target_f.append(0)   #0target_fgate_lst
            count_now=2*count_now-count_last   #
            count_last=count_now    #
            x=list(high_item_set.difference(current_item_set))[0] #
            current_item_set.add(x)      #x
            lst2,current_f,target_f2=syn_mon(x,num_qubit)  #current_f   
            gate_lst.extend(lst2)    ##lst2
            target_f.extend(target_f2)
            count_now=count_now+degree-1
            high_mat.append(search_item_lst(x+1,high_item_lst))  #search_item_lstxx+1
            high_index.append(count_now)
            if len(current_item_set)==len_high:     #
                gate_lst=gate_restore_lst(gate_lst,count_last,count_now)
                gate_lst,f=add_cnot(high_mat,high_index,f,gate_lst,high_item_lst,target_f,num_qubit)
                gate_lst=gate_lst_opt(gate_lst) #
                return gate_lst,f             #
        else:
            gate_lst.append(max_ijk)
            count_now=count_now+1
            i,j,k=max_ijk[0],max_ijk[1],max_ijk[2]
            current_f[k]=toffoli_f(current_f[i],current_f[j],current_f[k])
            target_f.append(current_f[k])




from sympy import * 
import math
import random
import copy
x= IndexedBase('x')

###
def multi_step_syn_opt(f,num_qubit):  
    degree=degree_rev(f)   #
    gate_lst=[]          #
    while degree>2:
        lst,f=elim_onedegree(f,num_qubit)
        gate_lst.extend(lst)
        degree=degree_rev(f)
    lst2=syn_low(f,num_qubit)
    gate_lst.extend(lst2)
    gate_lst=gate_lst_opt(gate_lst)
    return gate_lst





import numpy as np
from qiskit import *
from qiskit_aer import AerSimulator
num_qubit=5
len_f=5
q1=QuantumRegister(num_qubit,'a')
q2=QuantumRegister(len_f,'b')
c1 = ClassicalRegister(len_f)

qc=QuantumCircuit(q1,q2,c1)




##
# qc.x(q1[0])
# qc.x(q1[1])
# qc.x(q1[2])
# qc.x(q1[3])
# qc.x(q1[4])

s='040B1F141A1509021B0508121D03061C1E13070E000D1118100C0119160A0F17'  
f=str_to_fun(s)
print(f)

lst=multi_step_syn_opt(f,num_qubit)
print(lst)

lst_to_qc(lst,num_qubit)
for i in range(len_f):
    qc.measure(q2[i],c1[i])

qc.draw('mpl')
```
